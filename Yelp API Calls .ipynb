```python
from yelpapi import YelpAPI
import json
import pandas as pd
```

```python
with open('/Users/lbodden/.secret/yelp_api.json') as f: #change the path to match YOUR path!!
    login = json.load(f)
login.keys()
```

```python
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
pizza_0 = yelp_api.search_query(location='Bronx, NY',
                                       term='Pizza')
type(pizza_0)
pizza_0.keys()
```

```python
pizza0 = pd.DataFrame(pizza_0['businesses'])
pizza0.head(2)
```

```python
pizza_0['total']
```

```python
pizza_20 = yelp_api.search_query(location='Bronx, NY',
                                       term='Pizza',
                               offset = 20)
```

```python
pizza20 = pd.DataFrame(pizza_20['businesses'])
pizza20.head(2)
```

```python
#now concatenate all pages
all_pizza = pd.concat([pizza0, pizza20],
                      ignore_index=True)
display(all_pizza.head(3), all_pizza.tail(3))
```
